# 数値データを読み込んでグラフを描画する

以下では、スペースで区切られた数値が複数行書き込まれたテキストファイル "rand.dat" を、NumPyを利用して読み込み、Matplotlibを利用してグラフにする。

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("./rand.dat")
       # NumPyのloadtxt関数を用いて数値テキストファイルをNumPy配列に変換
index = data[:,0]      # スライスして第0列目だけ取り出す
random = data[:,1]     # スライスして第1列目だけ取り出す
plt.plot(index,random,"o-") # Matplotlibによるグラフの描画
plt.show()
```

### NumPyの関数やMatplotlibの機能を利用することで、さらに詳しい解析が行える。

```python
print(np.average(random)) # 平均値を求める
print(np.std(random))     # 標準偏差を求める
```

```python
x=np.linspace(random.min(),random.max(),100)
plt.hist(random,bins=x,density=True,label="Histogram")
plt.plot(x,np.exp(-x**2/2)/np.sqrt(2*np.pi),label="Gaussian distribution")
plt.xlabel("Value of random number")
plt.ylabel("Probability distribution function (PDF)")
plt.legend()
plt.show()
```
