# データの読み込み・テキスト処理

## 各行の列数が等しい数値が並んだテキストデータの読み込み：np.loadtxt

```python
import numpy as np
import matplotlib.pyplot as plt

# np.loadtxtによるテキストデータの読み込み
data = np.loadtxt("testdata01.dat")
t = data[:,0]
x = data[:,1]
y = data[:,2]
z = data[:,3]

# 2次元プロット
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(t,x,"o-",label="x(t)")
ax.plot(t,y,"x-",label="y(t)")
ax.set_xlabel("t")
ax.legend()
plt.show()

# 3次元プロット
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot(x,y,z,"o-")
ax.view_init(elev=30,azim=45)
plt.show()
```
