# Newton Raphson法の誤差の収束性

## 単根の場合
$$
\left|e_n\right| = \frac{1}{2}\left|\frac{f''(x^\star)}{f'(x^\star)}\right|\left|e_{n-1}\right|^2
$$

例題: $f(x) = x^2 -a = 0$ with $a=1000000$

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("error_single_root.dat")
it = data[:,0]
x = data[:,1]
error = data[:,2]

fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(it,x,"o-",label="$x_n$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Approximate solution $x_n$")
ax.set_yscale("log")
ax.legend()
plt.show()

fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(it,error,"o-",label="$e_n$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Error $e_n$")
ax.set_yscale("log")
ax.legend()
plt.show()
```

誤差の収束性の確認
$$
\left|e_n\right| = \frac{1}{2}\left|\frac{f''(x^\star)}{f'(x^\star)}\right|\left|e_{n-1}\right|^2
$$
（単根の場合）

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(it,error,"o-",label="$e_n$")
ax.plot(it[1:],error[:-1]**2/2000,"x-",label=r"$\frac{1}{2}\left|\frac{f''(x^\star)}{f'(x^\star)}\right|\left|e_{n-1}\right|^2$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Error $e_n$")
ax.set_yscale("log")
ax.legend()
plt.show()
```

## 重根の場合
$$
\left|e_n\right| = \frac{1}{2}\left|e_{n-1}\right|
$$

例題: $f(x) = (x - \sqrt{a})^2 = 0$ with $a=1000000$

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("error_double_root.dat")
it = data[:,0]
x = data[:,1]
error = data[:,2]

fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(it,x,"o-",label="$x_n$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Approximate solution $x_n$")
ax.set_yscale("log")
ax.legend()
plt.show()

fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(it,error,"o-",label="$e_n$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Error $e_n$")
ax.set_yscale("log")
ax.legend()
plt.show()
```

誤差の収束性の確認
$$
\left|e_n\right| = \frac{1}{2}\left|e_{n-1}\right|
$$
（重根の場合）

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
ax.plot(it,error,"o-",label="$e_n$")
ax.plot(it[1:],error[:-1]/2,"x-",label=r"$\frac{1}{2}\left|e_{n-1}\right|^2$")
ax.set_xlabel(r"Iteration $n$")
ax.set_ylabel(r"Error $e_n$")
ax.set_yscale("log")
ax.legend()
plt.show()
```
