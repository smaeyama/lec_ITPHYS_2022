```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# データの読み込み
data = np.loadtxt("trajectory.dat")
t = data[:,0]
x = data[:,1]
z = data[:,2]
vx = data[:,3]
vz = data[:,4]

# 軌道の描画
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,z)
ax.set_xlabel("x")
ax.set_ylabel("z")
plt.show()
```

```python
# 軌道のアニメーション表示
fig = plt.figure()
ax = fig.add_subplot()
def update_ax(i):
    ax.clear()
    ax.plot(x[:i+1],z[:i+1])
    ax.scatter(x[i],z[i],c="Grey")
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_xlim(x.min(),x.max())
    ax.set_ylim(0,1.1*z.max())
ani = FuncAnimation(fig, update_ax, frames=range(0,len(x),2), interval=50)
HTML(ani.to_jshtml())
```

```python
# エネルギー保存則の確認
kinetic = 0.5 * (vx**2 + vz**2)
potential = 9.8 * z
energy = kinetic + potential

# 軌道の描画
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(t,energy)
ax.set_xlabel("x")
ax.set_ylabel("Total energy")
ax.set_ylim(0,1.1*energy[0])
ax.grid()
plt.show()
```
