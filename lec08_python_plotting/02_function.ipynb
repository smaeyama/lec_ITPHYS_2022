# Pythonにおける関数定義
```
def 関数名(引数1, 引数2, ...):
    処理
    return 返り値1, 返り値2, ...
```

Pythonにおける関数の引数は参照渡しだが、**「引数は書き換えない。得たい値はすべて返り値として返す」**というシンプルな使い方を推奨。

## 例1.

````python
def swap(a, b):
    '''
    引数 a, b を入れ替えて、返り値として返す.
    '''
    return b, a

a=1.0
b=2.0
print(a,b)
a, b = swap(a,b)
print(a,b)

print("# 引数は書き換えない。得たい値はすべて返り値として返す。シンプルイズベスト！")
````

## 例2.

````python
import numpy as np
import matplotlib.pyplot as plt

def morlet_wavelet(t,omega=2*np.pi):
    phi = np.cos(omega*t)*np.exp(-0.5*t**2)/np.pi**0.25
    return phi
    
t=1.0
f = morlet_wavelet(t)
t_array = np.linspace(-4,4,200)
f_array = morlet_wavelet(t_array)

print(f)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t_array,f_array)
plt.show()
````
