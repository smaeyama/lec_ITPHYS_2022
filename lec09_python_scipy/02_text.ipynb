# データの読み込み・テキスト処理

## 文字・数値を含む一般的なテキストデータの場合

## np.loadtxtのキーワード引数usecolsを使った簡便な方法

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("Habble.dat", usecols=[1,2]) # 数値データの第1,2列のみ読み込む
velocity = data[:,0]
distance = data[:,1]

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(distance,velocity)
ax.set_xlabel(r"Distance $R$ [Mpc]")
ax.set_ylabel(r"Velocity $v$ [km/s]")
plt.show()
```

## ファイルオープンしてテキストデータを読み込む方法

```python
# テキストデータの読み込み
filename = "Habble.dat"
data = []  # 空のリストを用意し、必要な要素をdata.append()で追加していく。
with open(filename,mode="r") as f: # fという変数名でテキストファイルをオープン
    lines = f.readlines()          # fのすべての行を文字列として格納したリストを取得
    for line in lines:             # linesリストから1行ずつlineとして取り出す
        data.append(line.split())  # split: 空白の位置で文字列を分割する
print(data)

# 読み取ったデータの各要素を文字列リスト・Numpy配列に分ける
label = []
velocity = []
distance = []
for row in data:
    label.append(row[0])           # 最初の列はラベル情報（string）
    velocity.append(float(row[1])) # 次の列は後退速度(float)
    distance.append(float(row[2])) # その次の列は距離(float)
velocity = np.array(velocity) # 実数のリストであるvelocityやdistanceは解析しやすいようにNumpy配列に変換
distance = np.array(distance)

# グラフに描画
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(distance,velocity)
ax.set_xlabel(r"Distance $R$ [Mpc]")
ax.set_ylabel(r"Velocity $v$ [km/s]")
for i in range(len(label)):
    ax.text(distance[i]+0.3,velocity[i],label[i]) # 指定した座標に文字列を追記
plt.show()
```
