```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("error.dat")
n = data[:,0]
error_trapezoid = data[:,1]
error_simpson = data[:,2]

fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot()
ax.plot(n,np.abs(error_trapezoid),"o-",label="Trapezoid")
ax.plot(n,np.abs(error_simpson),"o-",label="Simpson")
ax.plot(n,0.2*n**-2,"--",label=r"$\propto h^2$")
ax.plot(n,0.1*n**-4,"--",label=r"$\propto h^4$")
ax.set_xlabel("Number of grid points $N$")
ax.set_ylabel("Error in numerical integration")
ax.set_xscale("log")
ax.set_yscale("log")
ax.legend()
fig.tight_layout()
plt.show()
```
