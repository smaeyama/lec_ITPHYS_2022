# グラフ描画パッケージ Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlibの考え方: グラフをオブジェクトとして取り扱う

## 最小のスクリプト: オブジェクトを陰に扱う記法

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## オブジェクトを陽に扱う記法

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()
```

## 1つの図に複数のグラフを描画する例
FigureオブジェクトとAxesオブジェクトの違いが分かりやすい

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x,np.sin(x))
ax1.plot(x,np.cos(x))
ax2 = fig.add_subplot(2,1,2)
ax2.plot(x,np.tan(x))
ax2.set_ylim(-10,10)
plt.show()
```
