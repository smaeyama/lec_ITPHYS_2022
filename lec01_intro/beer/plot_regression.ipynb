```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("df.csv",index_col=0,parse_dates=["month"])
df = df[df.index.year > 2015]
print(df)
```

```python
#線形回帰分析
slope, intercept, r_value, p_value, std_err = stats.linregress(df.temperature,df.beer)
def func(x,slope,intercept):
    return slope*x+intercept

#回帰直線のプロット
plt.scatter(df.temperature,df.beer)
plt.title("Linear regression between temperature and beer sales (Jan 2016 - Dec 2021)")
w=np.linspace(df.temperature.min(),df.temperature.max(),100)
plt.plot(w,func(w,slope,intercept),color="tab:orange")
plt.xlabel(r"Temperature [$^\circ$C]")
plt.ylabel(r"Beer sales [x10,000 Cases]")
plt.show()
print("Slope of the linear regression=",slope)
print("Intercept of the linear regression=",intercept)
print("Correlation coefficient R=",r_value)
print("Two-sided p-value for a hypothesis test whose null hypothesis is that the slope is zero. P=",p_value)
print("Standard error of the estimate=",std_err)

print("### Rough rule of thumb:                      ###\n"
      "###   Strong correlation ... |R|>0.7          ###\n"
      "###   Rejection of null hypothesis ... P<0.05 ###")
```
